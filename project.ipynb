```python
import pandas as pd
import numpy
import sklearn
```

```python
def applyOneHotEncoder(oneHotEncoder,dataframe,*coloumns):
    for coloumn in coloumns:
        encoded_features = oneHotEncoder.fit_transform(dataframe[[coloumn]])
        encoded_x = pd.DataFrame(encoded_features.toarray(), columns=oneHotEncoder.get_feature_names_out([coloumn]))
        dataframe = pd.concat([dataframe, encoded_x], axis=1)
    return dataframe
```

```python
attendance = pd.read_csv('attendence_results.csv')
leaves=pd.read_csv('leaves_result.csv')
salary=pd.read_csv('salary_results.csv')
employee=pd.read_csv('employee_preprocess_200487B.csv')
```

```python
emp_model = employee[["Employee_No","Gender","Marital_Status","Designation_ID","Status","Employment_Category","Employment_Type","Year_of_Birth","Date_Joined"]]
```

```python
emp_model["Year_joined"]=pd.to_datetime(emp_model["Date_Joined"]).dt.year
```

```python
emp_model
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oneHotEncoder = OneHotEncoder()
```

```python
emp_model=applyOneHotEncoder(oneHotEncoder,emp_model,"Gender","Marital_Status","Employment_Category","Employment_Type")
```

```python
emp_model=emp_model.drop(["Gender","Marital_Status","Employment_Category","Employment_Type","Status","Date_Joined"],axis=1)
```

```python
attendance_model=attendance[["Employee_No","isLate%","isEarly%","isHoliday%"]]
```

```python
emp_model=pd.merge(emp_model,attendance_model,on="Employee_No",how="left")
```

```python
emp_model
```

```python
salary=salary.drop("Unnamed: 0",axis=1)
```

```python
salary
```

```python
emp_model=pd.merge(emp_model,salary,on="Employee_No",how="left")
```

```python
emp_model=emp_model.fillna(0)
```

```python
from sklearn.cluster import DBSCAN,AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
```

```python
emp_model_to_cluster=emp_model.drop(["Employee_No"],axis=1)
```

```python
def cluster(model,data):
    model.fit(data)
    labels=model.labels_
    score = silhouette_score(data, labels)
    return score,labels
```

```python
score,lables = cluster(AgglomerativeClustering(n_clusters=4),emp_model_to_cluster)
```

```python
print(score)
```

```python
employee["cluser"]=lables
```

```python
results=employee.groupby(["cluser","Status"])["Employee_No"].count().reset_index()
```
